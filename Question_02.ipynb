```python
import pandas as pd
import re

# Load the CSV files
qoura_df = pd.read_csv("/content/drive/MyDrive/Quera/qoura_questions.csv")
shereno_df = pd.read_csv("/content/drive/MyDrive/Quera/shereno.csv")

# Function to count digits in a sentence
def count_digits(sentence):
    digits = re.findall(r'\d', sentence)
    return len(digits)

# Calculate digits for each question in qoura_questions.csv
qoura_df['num_digits'] = qoura_df['question'].apply(count_digits)

# Calculate digits for each poem in shereno.csv
shereno_df['num_digits'] = shereno_df['Poem'].apply(count_digits)

# Print the results
for index, row in qoura_df.iterrows():
    print(f"Question {index + 1}: {row['question']} has {row['num_digits']} digits.")

for index, row in shereno_df.iterrows():
    print(f"Poem {index + 1}: {row['Poem']} has {row['num_digits']} digits.")
```
