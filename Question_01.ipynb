```python
import pandas as pd
import re

# Load the CSV file
df = pd.read_csv("/content/drive/MyDrive/Quera/qoura_questions.csv")

# Function to count unique words in a sentence
def count_unique_words(sentence):
    words = re.findall(r'\b\w+\b', sentence.lower())
    return len(set(words))

# Calculate unique words for each question
df['unique_words'] = df['question'].apply(count_unique_words)

# Print the results
for index, row in df.iterrows():
    print(f"Question {index + 1}: {row['question']} has {row['unique_words']} unique words.")
```
