```python
import pandas as pd
import re

# Load the CSV file
shereno_df = pd.read_csv("/content/drive/MyDrive/Quera/shereno.csv")

# Load the stopwords from stopwords.txt
with open("/content/drive/MyDrive/Quera/stopwords.txt", "r") as f:
    stopwords = set(f.read().splitlines())

# Function to count stopwords in a sentence
def count_stopwords(sentence):
    words = re.findall(r'\b\w+\b', sentence.lower())
    return len(set(words) & stopwords)

# Calculate stopwords for each poem
shereno_df['num_stopwords'] = shereno_df['Poem'].apply(count_stopwords)

# Print the results
for index, row in shereno_df.iterrows():
    print(f"Poem {index + 1}: {row['Poem']} has {row['num_stopwords']} stopwords.")
```
